```python
from ultralytics import YOLO
```

```python
model = YOLO("C:/Users/91897/OneDrive/Desktop/cavity/yolov8_caries_detector/runs/detect/train4/weights/best.pt")
# model = YOLO("best.pt")
```

```python
img_path="C:\Users\91897\OneDrive\Desktop\Cavity Detection\yolov11_caries_detector\img\without_cavity\WhatsApp Image 2025-01-07 at 14.17.19_338b8bd3.jpg"
```

```python
results = model(img_path)
```

```python
boxes = results[0].boxes
len(boxes)
```

```python
caries_boxes = []
for box in boxes:
    class_name = results[0].names[box.cls.numpy()[0]]
    if class_name != "Caries":
        continue
    caries_boxes.append([int(v) for v in box.xyxy.numpy()[0]])
```

```python
from PIL import Image,ImageDraw
img = Image.open(img_path)
draw = ImageDraw.Draw(img)
for box in caries_boxes:
    draw.rectangle(box,None,"#00FF00",3)
img
```
